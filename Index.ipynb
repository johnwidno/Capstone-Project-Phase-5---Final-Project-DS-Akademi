# Libraries & Data Importation

```python
import pandas as pd
donors= pd.read_csv('Dataset/donors.csv')
campaigns= pd.read_csv('Dataset/campaigns.csv')
dons= pd.read_csv('Dataset/donations_linked.csv')
engagement_history= pd.read_csv('Dataset/engagement_history.csv')
engagement_outcomes =pd.read_csv('Dataset/engagement_outcomes.csv')
impact= pd.read_csv('Dataset/impact.csv')
```

### Merge all Tables

```python
Donors_dons= pd.merge(donors,dons , on='DonorID', how='left')
Donors_dons_Campaigns= pd.merge(Donors_dons, campaigns, on='CampaignID', how='left')
Donors_dons_Campaigns_engagement_history =pd.merge(Donors_dons_Campaigns, engagement_history, on='DonorID',  how='left')
Donors_dons_Campaigns_engagement_history_impacts =pd.merge(Donors_dons_Campaigns_engagement_history,impact, on ='CampaignID', how='left')
```

```python
df=Donors_dons_Campaigns_engagement_history_impacts
```

## Deal with Duplicates Values

```python
duplicates = df[df.duplicated()]
print("Duplicate rows:", duplicates.shape[0])
```

```python
duplicates.head()
```

```python
df = df.drop_duplicates(subset=['DonationID'])
```

```python
df.isnull().sum().sort_values(ascending=False)
```

## Find and drop the row with nan values

```python
df[df.isnull().any(axis=1)]
df = df.dropna(subset=['DonationID'])
```
